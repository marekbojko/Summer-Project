The set-up: Agents $\{1,2,...,N\}$ are placed in random environment of $\{1,2,...,K\}$ arms of a multi-armed bandit. In the simulations below, each arm yields a Bernoulli reward unless otherwise stated. Agents are placed in an all-to-all multi directed graph G, where the weight on a directed edge $(i,j)$ is the probability that agent $i$ shares her last observed pay-off with agent $j$. Agent's $i$'s fitness is represented as 

$$F(i) = \frac{\sum^{T}_{t=1}r_{it}}{1/N\sum_{j=1}^{N}\sum_{t=1}^{T}r_{jt}}$$

After selecting an arm and sharing information, agent $i$ changes her strategy randomly by adding nosie $\epsilon$ to her propensity to share with every other agent with $\epsilon \sim \mathcal{N}(0,\sigma_{\epsilon})$, where $\sigma_{\epsilon} = \frac{1}{F(i)}$.

## Initial embedding on a cycle

We place 50 agents in a field of 250 arms. Initially run in 10 different environments with constant strategy (constant propensity to share) 300 times, retrieve accumulated payoff. Then run 500 times with evolving strategies, agents change their strategy randomly (as described above) every 10 time steps. The environment changes only prior to the actual game.

```python
S = Simulation(50, 250, bernoulli_arms, discounted_thompson, 10, const=False)

S.plot_data()
```

We place 50 agents in a field of 250 arms. Initially run 3,000 times with constant strategies. Then run 500 times with evolving strategies. The bandits are 'restless' - at every 10 steps the expected payoff of every arm changes by adding the noise:

$$\epsilon \sim \mathcal{N}(0,0.05)$$

to their expected pay-off. Agents anticipate this and discount their priors by multiplying the information they posses about every arm by a constant discount factor $\gamma=0.9$. The environment keeps changing every 10 time steps throughout the simulation.

```python
P = Simulation(50, 250, bernoulli_arms, discounted_thompson,10,const=False)

P.plot_data("game_restless_bandits")
```
