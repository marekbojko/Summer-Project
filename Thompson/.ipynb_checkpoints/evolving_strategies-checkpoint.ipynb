```python
# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(-1, 2), ylim=(-1, 2))
particles_d, = ax.plot([], [], 'bo', ms=2)
particles_c, = ax.plot([], [], 'ro', ms=2)
#parents, = ax.plot([], [], 'ro', ms=4)

# initialization function: plot the background of each frame
def init():
    particles_c.set_data([], [])
    particles_d.set_data([], [])
    #parents.set_data([], [])
    return particles_c,particles_d

# animation function.  This is called sequentially
def animate(i):
    global locs,loc_c,loc_d
    x_c = [z[0] for z in loc_c[i]]
    y_c = [z[1] for z in loc_c[i]]
    particles_c.set_data(x_c,y_c)
    x_d = [z[0] for z in loc_d[i]]
    y_d = [z[1] for z in loc_d[i]]
    particles_d.set_data(x_d,y_d)
    return particles_c,particles_d
```

### Simulation 1
- 1000 agents with random upper bound p on pair-wise propensity to share placed randomly into a $[0,1]^2$ square
- The strategy p evolves with each generation, an offspring inherits p from the parent with random factor N(0,0.05)
- Agents first explore the bandits and accumulate rewards for 1000 time steps, then the environment changes randomly and they forget the information obtained but keep accumulated payoffs; this happens 6 times before the selection
- At the end of a generation, roulette-wheel algorithm selects parents to reproduce; offsprings are placed into the environment to the location $(x_p+N(0,0.01),y_p+N(0,0.01)$ where $(x_p,y_p)$ are the coordinates of the parents
- There are 100 generations in this simulation

```python
PD1 = Simulation_PD(100, 100, bernoulli_arms, discounted_thompson, 100, False, True, 1)
```

```python
locs,loc_c,loc_d = PD1.plot_data_mult_gen(True, 0.01)
```

```python
# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init,
                               frames=100, interval=100, blit=True)
rc('animation', html='html5')
anim
```
